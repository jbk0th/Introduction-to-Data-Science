```python

import pandas as pd
import numpy as np
import os

pd.set_option('max_rows', 300)
os.chdir('C:\\Users\\Jason\\Introduction-to-Data-Science') #changes the directory of the script into the correct one to access the 'Energy Indicators file
os.getcwd() #shows current working directory
#df = pd.read_excel('C:\\Users\\Jason\\Introduction-to-Data-Science\\Energy Indicators.xls')
	#the above method works as well for reading in the file with the full path name
columns=['0','1','Country','Energy Supply','Energy Supply per Capita','% Renewable']
   
energy = pd.read_excel('Energy Indicators.xls', header=None, skiprows=18, skip_footer=38, names=columns)

energy.drop(['0','1'],1,inplace=True)
#easy way to remove specific labeled cols/rows, pass in the strings and, 0 for rows, 1 for dropping columns
#if you want to reassign to a new df just create new var and assign, if not just use method , w/ inplace=True

energy['Country'].replace(['\d+'],'', inplace=True, regex=True)
#remove all #s from any country name (\d=[0-9] in Regex), and retain just country name

energy = energy.replace(['^\.+'], np.nan,regex=True) 
#replaced all '...' of any length with NaN

energy['Energy Supply'] = energy['Energy Supply']*(1000000) 
#changed PJ to gigajoules

country_corrections = {'Republic of Korea':'South Korea','United States of America':'United States','United Kingdom of Great Britain and Northern Ireland':'United Kingdom','China, Hong Kong Special Administrative Region':'Hong Kong'}
energy.replace(country_corrections,inplace=True)

energy['Country'] = energy['Country'].apply(lambda x: str(x).split('(')[0]) 
#fixed all countries with () | find any entry in that column with parantheses and first list elem

energy['Country'] = energy['Country'].str.strip()
#remove any whitespace either leading or trailing from the regex replace function, or the split ( function

energy
```

```python
### GDP dataset formatting
GDP = pd.read_csv('world_bank.csv',header=0,skiprows=4)
#supplied correct header row
country_corrections={'Korea, Rep.':'South Korea','Iran, Islamic Rep.':'Iran','Hong Kong SAR, China':'Hong Kong'}
GDP.replace(country_corrections,inplace=True)
#replaced all the country name with corrected versions, inplace
GDP = GDP.rename(columns={'Country Name':'Country'})
#renamed 'Country Name' to 'Country' to be able to merge common country names for new df
```

```python
### ScinEm

ScimEn = pd.read_excel('scimagojr-3.xlsx')
# read data into dataframe, already formatted very nicely
```

```python
country_info_intersect = pd.merge(energy, GDP, how='inner', on='Country').merge(ScimEn, how='inner', on='Country')

country_info = country_info_intersect[country_info_intersect['Rank'] <= 15]
columns = ['Country','Rank','Documents','Citable documents','Citations','Self-citations','Citations per document','H index','Energy Supply','Energy Supply per Capita','% Renewable','2006','2007','2008','2009','2010','2011','2012','2013','2014','2015']
country_info = country_info[columns]
country_info.set_index('Country',inplace=True)
country_info
```

## Week_3 #10

```python
### Create a new column with a 1 if the country's % Renewable value is at or above the median for all countries in the top15, and
### a 0 if the country's % Renewable value is below the median
```

```python
Top15['HighRenew'] = 0
med_renew = Top15['% Renewable'].median()
Top15.loc[Top15['% Renewable'] >= (med_renew), 'HighRenew'] = 1
# Very important lesson here was to use the loc/iloc when querying things otherwise, can do 1 followed by the other
# and will not be sure whether a view or a copy is returned, here becuase all in one op know its a view
# and can be sure the assignment will affect you're real df
Top15.sort_values('Rank',inplace=True) # sorted df inplace to be able to call the 'HighRenew' series from the df sorted the countries 
Top15['HighRenew']
```

```python
### Group countries by the provided df, the create a df that displays the sample size (number of countries in ceach continent bin)
### and the sum, mean, ans std deviation for the estimated population of each country
```
